2.   Типи структурованих даних:

У програмуванні та аналізі даних існує кілька основних типів структурованих даних. Ось найпоширеніші:


1.    **Числові (Numerical)**: Дані, що виражаються числами.
*   **Дискретні (Discrete)**: Приймають цілі значення (наприклад, кількість працівників, кількість відвідувачів).
*   **Неперервні (Continuous)**: Можуть приймати будь-яке значення в межах діапазону (наприклад, заробітна плата, температура, вага).
2.  **Категорійні (Categorical)**: Дані, що описують якісні характеристики, які можна розбити на групи (категорії).
*   **Номінальні (Nominal)**: Категорії без впорядкованості (наприклад, вид економічної діяльності, стать, країна).
*   **Порядкові (Ordinal):** Категорії з певним порядком або ранґом (наприклад, рівень освіти: "середня", "бакалавр", "магістр"; оцінка задоволення: "погано", "добре", "відмінно").
3.   **Текстовий (Text/String)**: Рядки символів, що представляють текст (наприклад, назви, описи, коментарі).
4.   **Дата/Час (Datetime)**: Спеціальний тип даних для представлення дат і часу.
5.   **Булеві (Boolean)**: Дані, що приймають лише два значення: True (Істина) або False (Хиба).

```python
selected_features = ['2010', '2015', '2020', '2024']
```

```python
def calculate_statistics(data):
    stats_dict = {}

    # a. Середнє
    stats_dict['Середнє'] = np.mean(data)

    # b. Медіана
    stats_dict['Медіана'] = np.median(data)

    # c. Посічене середнє (15% для непарного варіанту)
    stats_dict['Посічене середнє (15%)'] = stats.trim_mean(data, 0.15)

    # d. Дисперсія
    stats_dict['Дисперсія'] = np.var(data, ddof=0)  # ddof=0 для дисперсії популяції

    # e. Стандартне відхилення
    stats_dict['Стандартне відхилення'] = np.std(data, ddof=0)

    # f. Середнє абсолютне відхилення
    stats_dict['Середнє абсолютне відхилення'] = np.mean(np.abs(data - np.mean(data)))

    # g. Мінімальне та максимальне значення
    stats_dict['Мінімальне значення'] = np.min(data)
    stats_dict['Максимальне значення'] = np.max(data)

    # h. Розмах
    stats_dict['Розмах'] = np.max(data) - np.min(data)

    # i. Процентиль (95-й для непарного варіанту)
    stats_dict['95-й процентиль'] = np.percentile(data, 95)

    # j. Міжквартильний розмах
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    stats_dict['Міжквартильний розмах'] = Q3 - Q1

    return stats_dict
```

Розраховую статистики для кожної ознаки.

```python
# Створюємо словник для зберігання результатів
all_stats = {}

for feature in selected_features:
    feature_data = df[feature].dropna()  # Видаляємо NaN значення
    all_stats[feature] = calculate_statistics(feature_data)
    print(f"   Статистики для {feature} року:")
    print(f"   - Середнє: {all_stats[feature]['Середнє']:.2f}")
    print(f"   - Медіана: {all_stats[feature]['Медіана']:.2f}")
    print(f"   - Стандартне відхилення: {all_stats[feature]['Стандартне відхилення']:.2f}")
    print()
```

Формули та спососби розрахунку.

```python
formulas = {
    'Середнє': 'μ = (Σxᵢ) / n',
    'Медіана': 'Значення, що ділить впорядковану вибірку на дві рівні частини',
    'Посічене середнє (15%)': 'Середнє після видалення 15% найменших та 15% найбільших значень',
    'Дисперсія': 'σ² = Σ(xᵢ - μ)² / n',
    'Стандартне відхилення': 'σ = √σ²',
    'Середнє абсолютне відхилення': 'MAD = Σ|xᵢ - μ| / n',
    'Мінімальне значення': 'min(x)',
    'Максимальне значення': 'max(x)',
    'Розмах': 'max(x) - min(x)',
    '95-й процентиль': 'Значення, нижче якого знаходиться 95% спостережень',
    'Міжквартильний розмах': 'IQR = Q₃ - Q₁'
}
```

Створюю таблицю результатів.

```python
results_data = []
indicators = list(formulas.keys())

for indicator in indicators:
    row = [indicator, formulas[indicator]]
    for feature in selected_features:
        if indicator in all_stats[feature]:
            value = all_stats[feature][indicator]
            # Форматуємо числа для кращого відображення
            if isinstance(value, (int, float)):
                row.append(f"{value:.2f}")
            else:
                row.append(str(value))
        else:
            row.append("—")
    results_data.append(row)
```

```python
# Створюємо DataFrame
results_df = pd.DataFrame(results_data,
                         columns=['Показник', 'Формула / спосіб розрахунку'] + selected_features)

# Відображаємо таблицю
print(results_df.to_string(index=False))
```
