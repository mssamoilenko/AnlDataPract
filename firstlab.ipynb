```python
import pandas as pd
```

```python
df = pd.read_excel("szpshp_ed_rik.xlsx", header=None)
```

```python
header_row = df[df.iloc[:,0].astype(str).str.contains("Вид діяльності", na=False)].index[0]
```

```python
df.columns = df.iloc[header_row]
df = df.iloc[header_row+1:]
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df = df.dropna(axis=0, how='all')
```

```python
df = df.reset_index(drop=True)
```

```python
df.rename(columns={"Вид діяльності": "Activity_UA"}, inplace=True)
```
